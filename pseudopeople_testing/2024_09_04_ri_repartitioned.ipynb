```python
def is_categorical(array_like):
    return array_like.dtype.name == 'category'

import glob
import pandas as pd
from tqdm.notebook import tqdm
from pathlib import Path

for dataset in glob.glob('/mnt/team/simulation_science/pub/models/vivarium_census_prl_synth_pop/results/release_02_yellow/full_data/united_states_of_america/2023_08_21_16_35_27/final_results/2023_08_31_15_58_01/states/pseudopeople_simulated_population_rhode_island_2_0_0/*'):
    if "CHANGELOG.rst" in dataset:
        continue
    dataset_name = Path(dataset).name
    print(dataset_name)

    shard_files = glob.glob(dataset + '/*.parquet')
    if len(shard_files) == 0:
        print('No shards found')
        continue

    df = []
    for shard_file in tqdm(shard_files):
        df_part = pd.read_parquet(shard_file)
        for col in df_part.columns:
            if is_categorical(df_part[col]):
                df_part[col] = df_part[col].cat.remove_unused_categories()
        df.append(df_part)

    df = pd.concat(df)

    year_cols = [
        c for c in ['year', 'tax_year'] if c in df.columns
    ]
    assert len(year_cols) <= 1
    if 'survey_date' in df.columns:
        df = df.sort_values('survey_date')
    elif 'event_date' in df.columns:
        df = df.sort_values('event_date')
    
    directory = dataset.replace('pseudopeople_simulated_population_rhode_island_2_0_0', 'pseudopeople_simulated_population_rhode_island_experimental_repartition')
    Path(directory).mkdir(parents=True, exist_ok=True)
    df.to_parquet(
        f'{directory}/{dataset_name}_all.parquet',
        partition_cols=year_cols,
    )
```
