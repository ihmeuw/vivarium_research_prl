## Check for trailing whitespace after numbers in street number in 2020 census

Looks like there's none. Good.

```python
nonnumeric = census.loc[~(census.street_number.isna() | census.street_number.str.isdigit()), 'street_number']
nonnumeric
```

```python
# Find strings that contain whitespace
nonnumeric.loc[nonnumeric.str.contains('\s')]
```

```python
# Check if any strings begin or end with whitespace - Nope! Good.
nonnumeric.str.contains('$\s|\s^').any()
```

```python
int_cols = ['age', 'street_number', 'unit_number']
for col in int_cols:
    print(col, census[col].str.contains('$\s|\s^').any())
```
